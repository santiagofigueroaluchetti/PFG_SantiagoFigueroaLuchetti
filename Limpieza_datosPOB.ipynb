```python
import pandas as pd
```

```python
# Leer el archivo CSV especificando que ciertas columnas deben ser leídas como texto (string)
df = pd.read_csv('conjunto_de_datos_ageb_urbana_09_cpv2020.csv', dtype=str)
df.head(20)
```

```python
# Concatenar las variables en una nueva columna llamada 'CODE'
df['CODE'] = df['ENTIDAD'] + df['MUN'] + df['LOC'] + df['AGEB'] + df['MZA']
result = df[['CODE']]
result.head(150)
```

```python
df.head()
```

```python
import fiona
import geopandas as gpd
# Cargar los shapefiles
gdf3 = gpd.read_file('RPY_Manzanas_Sample.shp')
gdf3d = gpd.read_file('RPY_Manzanas_Sample.dbf')
```

```python
# Mostrar las etiquetas (es decir, los atributos)
print(gdf3.columns)
gdf3.head(30)
```

```python
# Realizar el merge con 'inner' para eliminar los CODE que no coinciden con CVEGEO
merged_df = pd.merge(df, gdf3, left_on='CODE', right_on='CVEGEO', how='inner')

# Verificar el resultado
print(merged_df.head())
print(f"Total de filas después del merge: {len(merged_df)}")
```

```python
merged_df.columns.tolist()
```

```python
# Eliminar las columnas 'a', 'b', 'c' del DataFrame
cleaned_df = merged_df.drop(columns=['CVEGEO','ENTIDAD', 'MUN', 'LOC', 'NOM_LOC','AGEB','MZA','CVE_ENT', 'CVE_MUN','CVE_LOC','CVE_AGEB','CVE_MZA','AMBITO','TIPOMZA',])
cleaned_df = cleaned_df[['CODE'] + [col for col in cleaned_df.columns if col != 'CODE']]
cleaned_df.head(20)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Convertir el DataFrame en un GeoDataFrame
gdf_31001 = gpd.GeoDataFrame(cleaned_df, geometry='geometry')
# Definir un CRS (ejemplo, WGS 84)
gdf_31001.set_crs("EPSG:4326", inplace=True)
# Guardar el GeoDataFrame en un archivo Shapefile
gdf_31001.to_file('Reproyectada_datospublicos_SAMPLE_CDM.shp')
```
