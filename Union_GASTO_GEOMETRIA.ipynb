```python
import pandas as pd
```

```python
# Leer el archivo CSV especificando que ciertas columnas deben ser leídas como texto (string)
df = pd.read_excel('Tipología_de_gasto_sample.xlsx', dtype=str)
df.head(20)
```

```python
import fiona
import geopandas as gpd
# Cargar los shapefiles
gdf3 = gpd.read_file('RPY_Manzanas_Sample.shp')
gdf3d = gpd.read_file('RPY_Manzanas_Sample.dbf')
```

```python
# Mostrar las etiquetas (es decir, los atributos)
print(gdf3.columns)
gdf3.head(30)
```

```python
gdf3 = gdf3.rename(columns={"CVEGEO": "CODE"})
```

```python
df_gasto_filtrado = gdf3[gdf3["MICROCODE"] == "0901000010012"]
print(df_gasto_filtrado)
```

```python
# Realizar el merge de los dos DataFrames usando las columnas 'CODE' y 'CVEGEO'
merged_df = pd.merge(df, gdf3, left_on='MICROCODE_MUN', right_on='MICROCODE', how='inner')
merged_df.head()

```

```python
df_gasto_filtrado2 = merged_df[merged_df["MICROCODE"] == "0901000010012"]
df_gasto_filtrado2
```

```python
print(merged_df.columns)
```

```python
# Eliminar las columnas 'a', 'b', 'c' del DataFrame
cleaned_df = merged_df.drop(columns=['MICROCODE_MUN','CVE_ENT', 'CVE_MUN','CVE_LOC','CVE_AGEB','CVE_MZA','AMBITO','TIPOMZA'])
cleaned_df = cleaned_df[['CODE'] + [col for col in cleaned_df.columns if col != 'CODE']]
cleaned_df.head(20)
```
